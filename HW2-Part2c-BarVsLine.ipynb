```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load the csv file of Dow Jones Historical Prices
```

```python
colnames = ['date', 'open', 'high', 'low', 'close']
df=pd.read_csv("https://raw.githubusercontent.com/farha-neu/cs7295/master/DowJones-HistoricalPrices.csv",skiprows=[0],names=colnames,parse_dates=['date'])
```

```python
#Plot date on X-axis and high on Y-axis using line graph
```

```python
plt.figure(figsize=(10, 8))    
plt.plot(df['date'], df['high'])
plt.xlabel('Date', fontsize=12)
plt.ylabel('High', fontsize=12)
plt.ylim(ymin=18000)
plt.title('Dow Jones Historical Prices Per Day')
plt.show()
```

```python
#Plot date on X-axis and high on Y-axis using bar graph
```

```python
plt.figure(figsize=(10, 8))
plt.bar(df['date'], df['high'],align='center')
plt.xlabel('Date', fontsize=12)
plt.ylabel('High', fontsize=12)
plt.ylim(ymin=18000)
plt.title('Dow Jones Historical Prices Per Day')
plt.show()
```

```python
#Answer:
#I think line graph more effectively depicts stock market index per day. As there is a wide range of data, 
#repeated bars in the bar graph seems distracting. On the otherhand, line graph expresses all the relevant 
#information with clearer trend.
```
