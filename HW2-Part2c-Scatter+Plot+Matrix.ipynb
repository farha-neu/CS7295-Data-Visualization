```python
import numpy as np
import matplotlib.pyplot as plt
from pandas.plotting import scatter_matrix
import pandas as pd
```

```python
#Load 4 csv files of different stock markets
```

```python
colnames = ['date', 'open', 'high', 'low', 'close']
df=pd.read_csv("https://raw.githubusercontent.com/farha-neu/cs7295/master/DowJones-HistoricalPrices.csv",skiprows=[0],names=colnames,parse_dates=['date'])
nd=pd.read_csv("https://raw.githubusercontent.com/farha-neu/cs7295/master/NASDAQ-HistoricalPrices.csv",skiprows=[0],names=colnames,parse_dates=['date'])
ny=pd.read_csv("https://raw.githubusercontent.com/farha-neu/cs7295/master/NYSE-HistoricalPrices.csv",skiprows=[0],names=colnames,parse_dates=['date'])
sp=pd.read_csv("https://raw.githubusercontent.com/farha-neu/cs7295/master/SP500-HistoricalPrices.csv",skiprows=[0],names=colnames,parse_dates=['date'])
```

```python
# create a dataframe with closing prices of 4 stock markets
```

```python
data = pd.DataFrame()
data['df'] = df['close']
data['nd'] = nd['close']
data['ny'] = ny['close']
data['sp'] = sp['close']
data.columns = ['Dow Jones Close', 'NASDAQ Close', 'NYSE Close', 'SP500 Close']
```

```python
# Create a scatter matrix plot with 4 closing stock market prices
```

```python
Axes=scatter_matrix(data, alpha=0.6,figsize=(14, 14), diagonal='hist')
plt.suptitle('Scatter Matrix plot showing correlation between multiple stock market prices',size=16)
plt.show()
```

```python
#Answer:
#a. I used Scatter Plot Matrix(SPLOM) to represent relationships among closing stock prices of Dow Jones, 
# NASDAQ, NYSE, and SP500. Scatter plot is a good choice to find correlation between two quantitative variables and
# SPLOM represents relationship among several variables taken two at a time.
#b. From the graph, we see that there is a strong correlation between closing price of a stock market with other 3
# stock market closing prices.
```
