```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
from matplotlib.ticker import FuncFormatter
```

```python
#Load all the csv file of Dow Jones
```

```python
colnames = ['date', 'open', 'high', 'low', 'close']
df=pd.read_csv("https://raw.githubusercontent.com/farha-neu/cs7295/master/DowJones-HistoricalPrices.csv",skiprows=[0],names=colnames,parse_dates=['date'])
```

```python
#Finding the rate of change of stock market closing price from previous day
```

```python
df['ret'] = df['close'].apply(pd.to_numeric, errors='coerce').pct_change()*100
```

```python
plt.figure(figsize=(15, 8))
ax = plt.subplot(221)   
```

```python
#Remove the borders of top and bottom
```

```python
ax.spines["top"].set_visible(False)       
ax.spines["right"].set_visible(False)      
ax.get_xaxis().tick_bottom()    
ax.get_yaxis().tick_left() 
```

```python
#Plot date along X-axis and growth rate in Y-axis
```

```python
plt.plot(df['date'], df['ret'])
plt.xlabel('Date', fontsize=12)
plt.ylabel('Growth Rate in %', fontsize=12)
plt.axhline(0,lw=1)
plt.title('Dow Jones Historical Price growth rate per day')
plt.tight_layout(pad=0.4, w_pad=5, h_pad=5)
plt.show()
```

```python
#Answer:
# a.I have used a line graph to show the trend of growth rate of closing stock price of Dow Jones
# from previous day. As there is a wide range of data with regular interval, line graph shows a clearer trend. As 
# growth rates can be negtive, I have used a horizontal line along Y=0 so that 
# positive and negative growth rates can be easily distinguished,

#b. The index chart depicts date along X-axis and stock closing price growth rate from previous day along Y-axis.
# Overall, the change in growth rate fluctuates around 0%.
```
