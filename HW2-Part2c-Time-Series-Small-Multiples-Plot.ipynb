```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
```

```python
#Load all 4 csv files of stock markets
```

```python
colnames = ['date', 'open', 'high', 'low', 'close']
df=pd.read_csv("https://raw.githubusercontent.com/farha-neu/cs7295/master/DowJones-HistoricalPrices.csv",skiprows=[0],names=colnames,parse_dates=['date'])
nd=pd.read_csv("https://raw.githubusercontent.com/farha-neu/cs7295/master/NASDAQ-HistoricalPrices.csv",skiprows=[0],names=colnames,parse_dates=['date'])
ny=pd.read_csv("https://raw.githubusercontent.com/farha-neu/cs7295/master/NYSE-HistoricalPrices.csv",skiprows=[0],names=colnames,parse_dates=['date'])
sp=pd.read_csv("https://raw.githubusercontent.com/farha-neu/cs7295/master/SP500-HistoricalPrices.csv",skiprows=[0],names=colnames,parse_dates=['date'])
```

```python
#Plot date and highest price per day of Dow Jones 
```

```python
plt.figure(figsize=(15, 10))
ax = plt.subplot(221) 
ax.spines["top"].set_visible(False)       
ax.spines["right"].set_visible(False)
plt.plot(df['date'], df['high'])
plt.xlabel('Date', fontsize=12)
plt.ylabel('High', fontsize=12)
plt.ylim(ymin=0)
plt.ylim(ymax=27000)
plt.title('Dow Jones High Price Per Day',fontsize=14)
```

```python
#Plot date and highest price per day of NASDAQ
```

```python
ax2=plt.subplot(222)
ax2.spines["top"].set_visible(False)       
ax2.spines["right"].set_visible(False)
plt.plot(nd['date'], nd['high'])
plt.xlabel('Date', fontsize=12)
plt.ylabel('High', fontsize=12)
plt.ylim(ymin=0)
plt.ylim(ymax=27000)
plt.title('NASDAQ High Price Per Day',fontsize=14)
#plt.show()
```

```python
#Plot date and highest price per day of NYSE
```

```python
ax3=plt.subplot(223)
ax3.spines["top"].set_visible(False)       
ax3.spines["right"].set_visible(False)
plt.plot(ny['date'], ny['high'])
plt.xlabel('Date', fontsize=12)
plt.ylabel('High', fontsize=12)
plt.ylim(ymin=0)
plt.ylim(ymax=27000)
plt.title('NYSE High Price Per Day',fontsize=14)
```

```python
#Plot date and highest price per day of SP500
```

```python
ax4=plt.subplot(224)
ax4.spines["top"].set_visible(False)       
ax4.spines["right"].set_visible(False)
plt.plot(sp['date'], sp['high'])
plt.xlabel('Date', fontsize=12)
plt.ylabel('High', fontsize=12)
plt.ylim(ymin=0)
plt.ylim(ymax=27000)
plt.title('SP500 High Price Per Day',fontsize=14)
plt.tight_layout(pad=0.4, w_pad=5, h_pad=5)
plt.show()
```
