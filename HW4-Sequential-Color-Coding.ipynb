```python
import pandas as pd
import numpy as np
import matplotlib
import seaborn as sns
from matplotlib import cm
import matplotlib.pyplot as plt
```

```python
#Read data
df=pd.read_csv("https://raw.githubusercontent.com/farha-neu/cs7295/master/property_assesment.csv")

#Find the mean of gross tax and average assesed p
dd= df.groupby('ST_NAME', as_index=False).agg({"GROSS_TAX": "mean", "AV_TOTAL" : "mean"})
x = dd['ST_NAME']
y = dd['AV_TOTAL']
z = dd['GROSS_TAX']
```

```python
#set the size of figure
plt.figure(figsize=(30, 15)) 

#set font size
matplotlib.rc('font', size=18)
```

```python
#create colorbar with z values
colors = cm.YlGn(z / float(max(z)))

#map it to YlGn sequential colormap
plot = plt.scatter(z, z, c = z, cmap = 'YlGn')
plt.clf()
cbar = plt.colorbar(plot)
```

```python
#plot x and y values with bar color mapped to colorbar
bars= plt.bar(x,y,align='center',color = colors,edgecolor='black')

#set black border around each bar
for bar in bars:
    bar.set_edgecolor("black")
    bar.set_linewidth(1)
#provide labels
plt.xlabel('Street Name')
plt.ylabel('Average assessed property value')
plt.title('Average assessed property value and gross tax by street name- City of Boston Parcel Data',fontsize=22)
```

```python
#label for colorbar
cbar.ax.get_yaxis().labelpad = 15
cbar.ax.set_ylabel('Average Gross tax', rotation=270)
plt.xticks(rotation=90)
```

```python
plt.show()
```
